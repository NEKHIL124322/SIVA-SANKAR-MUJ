```python
import matplotlib.pyplot as plt

# Data points
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]

# Create a line plot
plt.plot(x, y, marker='o', linestyle='-', color='b', label="Sales")

# Adding labels and title
plt.xlabel("Days")
plt.ylabel("Revenue ($)")
plt.title("Daily Sales Trend")

# Show legend
plt.legend()

# Display the graph
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ["Apples", "Bananas", "Cherries", "Dates", "Elderberries"]
values = [50, 80, 60, 90, 40]

# Create a bar chart
plt.bar(categories, values, color='skyblue')

# Adding labels and title
plt.xlabel("Fruits")
plt.ylabel("Quantity Sold")
plt.title("Fruit Sales Data")

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
labels = ["Apples", "Bananas", "Cherries", "Dates"]
sizes = [40, 30, 20, 10]  # Percentage distribution
colors = ["red", "yellow", "pink", "brown"]

# Create pie chart
plt.pie(sizes, labels=labels, colors=colors, autopct="%1.1f%%", startangle=140)

# Adding title
plt.title("Fruit Sales Distribution")

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# Generate sample data
np.random.seed(42)  # For reproducibility
x = np.random.randint(1, 100, 20)
y = np.random.randint(1, 100, 20)
z = np.random.randint(1, 100, 20)

# Create a 3D figure
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

# Plot scatter points
ax.scatter(x, y, z, c=z, cmap='viridis', marker='o', s=50)

# Labels and title
ax.set_xlabel("X Axis")
ax.set_ylabel("Y Axis")
ax.set_zlabel("Z Axis")
ax.set_title("3D Scatter Plot")

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data points
x = [10, 20, 30, 40, 50, 60, 70, 80]
y = [5, 15, 7, 20, 10, 25, 30, 18]

# Create scatter plot
plt.scatter(x, y, color='blue', marker='o', s=100, label="Data Points")

# Adding labels and title
plt.xlabel("X-axis (Independent Variable)")
plt.ylabel("Y-axis (Dependent Variable)")
plt.title("Simple Scatter Plot")

# Show legend
plt.legend()

# Display the graph
plt.show()
```
